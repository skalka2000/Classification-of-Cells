```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140]).astype(int)
X_sets_grey=[]
for X_set in X_sets:
    X_set_grey=[]
    for X_Sample in X_set:
        X_Sample_grey=rgb2gray(X_Sample)
        X_set_grey.append(X_Sample_grey)
    X_sets_grey.append(X_set_grey)
```

```python
#transform  the 2d image reppresentations to 1d vector:
X_sets_1d=[]
for X_set in X_sets_grey:
    X_set_1d=[]
    for X_Sample in X_set:
        X_Sample_1d = X_Sample.reshape(-1)
        X_set_1d.append(X_Sample_1d)
    X_sets_1d.append(X_set_1d)
```

```python
# PCA
from sklearn.decomposition import IncrementalPCA #import PCA
X_ipca=[]
variance_ratios=[]
ipca = IncrementalPCA(n_components=5)
for X_set_1d in X_sets_1d:
    x_ipca = ipca.fit_transform(X_set_1d)
    X_ipca.append(x_ipca)
    variance_ratio = (ipca.singular_values_ ** 2) / np.sum(ipca.singular_values_ ** 2)
    variance_ratios.append(variance_ratio)
```

```python
X_ipca_single=[]
for x_ipca in X_ipca:
    x_ipca_single=[]
    for X_Sample in x_ipca:
        X_Sample_ipca=X_Sample[0]
        x_ipca_single.append(X_Sample_ipca)
    X_ipca_single.append(x_ipca_single)
```

```python
from skimage.feature import greycomatrix, greycoprops

dist = 5
angles = [0]
X_contrast=[]
X_energy=[]
X_homogeneity=[]
X_correlation=[]
for X_set_grey in X_sets_grey:
    x_contrast=[]
    x_energy=[]
    x_homogeneity=[]
    x_correlation=[]
    for X_Sample in X_set_grey:
        glcm = greycomatrix(X_Sample, [dist], angles, 256, symmetric=True, normed=True)
        # Compute GLCM features
        contrast = greycoprops(glcm, 'contrast')[0][0]
        energy = greycoprops(glcm, 'energy')[0][0]
        homogeneity = greycoprops(glcm, 'homogeneity')[0][0]
        correlation = greycoprops(glcm, 'correlation')[0][0]
        x_contrast.append(contrast)
        x_energy.append(energy)
        x_homogeneity.append(homogeneity)
        x_correlation.append(correlation)
    X_contrast.append(x_contrast)
    X_energy.append(x_energy)
    X_homogeneity.append(x_homogeneity)
    X_correlation.append(x_correlation)
```

```python
#!pip install torchvision
from torchvision.transforms import ToTensor
from torch.utils.data import Dataset, DataLoader


class CellImageDataset(Dataset):
    def __init__(self, images, labels):
        self.images = images
        self.labels = labels

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]
        image_tensor = ToTensor()(image)
        return image_tensor, label
```

```python
#create the dataloaders

train_dataset = CellImageDataset(list(X1),np.array(Y1_sum))
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
valid_dataset = CellImageDataset(list(X2),np.array(Y2_sum))
valid_loader = DataLoader(valid_dataset, batch_size=16, shuffle=False)
test_dataset = CellImageDataset(list(X3),np.array(Y3_sum))
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

```python
import torch
import torch.nn as nn

class CellCountingCNN(nn.Module):
    def __init__(self):
        super(CellCountingCNN, self).__init__()
        self.layer1=nn.Sequential(
        nn.Conv2d(3,16,kernel_size=5,stride=1,padding=2),
        nn.BatchNorm2d(16),
        nn.ReLU(),
        nn.MaxPool2d(kernel_size=2,stride=2))
        self.layer2=nn.Sequential(
        nn.Conv2d(16,32,kernel_size=5,stride=1,padding=2),
        nn.BatchNorm2d(32),
        nn.ReLU(),
        nn.MaxPool2d(kernel_size=2,stride=2))
        self.fc=nn.Linear(131072, 1)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.reshape(out.size(0),-1)
        out=self.fc(out)
        
        return out

model = CellCountingCNN().to('cpu')
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
# Train the model
num_epochs = 20
min_valid_loss=np.inf
epoch_train_losses=[]
epoch_valid_losses=[]
#epoch_patience=5;
for epoch in range(num_epochs):
    train_loss=0.0
    for i, (images, counts) in enumerate(train_loader):
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, counts.view(-1, 1).float())
        #loss.item()=float(loss.item())
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
        if (i+1) % 10 == 0:
            print('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}'
                  .format(epoch+1, num_epochs, i+1, len(train_dataset)//16, loss.item()))
    epoch_train_losses.append(train_loss)
    print('Epoch',epoch+1,'loss: ',train_loss) 
    valid_loss=0.0
    model.eval()
    with torch.no_grad():
        for i, (images_valid, counts_valid) in enumerate(valid_loader):
            outputs_valid = model(images_valid)
            loss_v=criterion(outputs_valid, counts_valid.view(-1, 1).float())
            valid_loss += loss_v
        #update the mininimum validation loss if current loss is lower 
        if valid_loss<min_valid_loss:
            min_valid_loss=valid_loss
            print('Updated validation loss for epoch',epoch+1)
            torch.save(model.state_dict(),'cnn.pth')
            not_improves=0
        else:
            not_improves=not_improves+1
#    if not_improves==epoch_patience:
#        break
    epoch_valid_losses.append(valid_loss)
    print('Current validation loss:', valid_loss)
```
